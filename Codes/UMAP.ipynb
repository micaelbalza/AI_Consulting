# Redução de Dimensionalidade com UMAP

## Introdução

O **UMAP (Uniform Manifold Approximation and Projection)** é uma técnica de redução de dimensionalidade que preserva tanto as relações **locais** quanto as **globais** dos dados, permitindo identificar agrupamentos e padrões de forma eficiente. Ele é amplamente utilizado para análise exploratória e visualização de dados complexos em 2D ou 3D.

### 1. Importar Bibliotecas e Dataset

```python
# Importar bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from umap.umap_ import UMAP
```

### 2. Carregar e Explorar o Dataset

```python
# Carregar o dataset Digits
digits = load_digits()
X, y = digits.data, digits.target

# Exibir informações sobre o dataset
print("Formato do dataset:", X.shape)
print("Número de classes:", len(np.unique(y)))

# Visualizar algumas imagens do dataset
fig, axes = plt.subplots(1, 5, figsize=(6, 3))
for i, ax in enumerate(axes):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f'Dígito: {y[i]}')
    ax.axis('off')
plt.suptitle("Exemplos do Dataset Digits")
plt.show()
```

```python
# (1797, 64) indica que há 1797 exemplos, cada um com 64 atributos.
# Os 64 atributos correspondem aos pixels de uma imagem 8x8, achatados em um vetor unidimensional.
```

### 3. Aplicar o UMAP

Configurações do UMAP:

- **n_neighbors=15**: Número de vizinhos considerados ao construir o grafo de proximidade.
- **n_components=2**: Reduz os dados para 2 dimensões.
- **min_dist=0.1**: Controla a dispersão dos pontos no espaço reduzido.
- **random_state=42**: Garante resultados reprodutíveis.

```python
# Configurar e aplicar o UMAP
umap = UMAP(n_neighbors=15, n_components=2, random_state=42, min_dist=0.1)
X_umap = umap.fit_transform(X)
```

### 4. Visualizar os Dados Reduzidos

```python
# Visualizar os dados reduzidos
plt.figure(figsize=(6, 4))
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='tab10', s=15)
plt.colorbar(scatter, label='Dígito')
plt.title("Redução de Dimensionalidade com UMAP")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.grid()
plt.show()
```

### Conclusão
O UMAP mostrou-se eficaz para projetar os dados do dataset Digits de um espaço de alta dimensionalidade (64 dimensões) para um espaço de 2 dimensões, preservando os agrupamentos das classes originais. Isso é especialmente útil para análise exploratória e visualização de padrões em datasets complexos.
